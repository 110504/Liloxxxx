```python
import matplotlib.pyplot as plt
# Data Financial Position Conventional Life Insurance
nama_akun = ['Deposito Berjangka', 'Surat Berharga yang Diterbitkan NRI', 'Saham', 'Obligasi Korporasi', 'Reksadana']
jumlah_investasi_januari_sampai_november_2023 = [392085892, 1654776597, 1639786334, 386824259, 1006131669]
# Membuat pie chart
plt.figure(figsize=(3, 3))
plt.pie(jumlah_investasi_januari_sampai_november_2023, labels=nama_akun, autopct='%1.1f%%', startangle=140)
# Menambahkan judul
plt.title('Data Financial Position Conventional Life Insurance')
# Menampilkan pie chart
plt.show()
```

Dari output diatas dapa disimpulkan bahwa Saham dan Surat Berharga yang Diterbitkan NRI merupakan dua instrumen investasi terbesar dalam portofolio asuransi jiwa konvensional dari kelima jenis instrumen investasi yang dianalisis. Masing-masing dengan persentase 32.3% dan 32.6%.
Reksadana dengan persentase 19.8% yang merupakan instrumen investasi berikutnya yang cukup signifikan. Sedangkan Instrumen investasi dengan persentase terkecil adalah deposito berjangka dan obligasi korporasi, masing-masing dengan persentase 7.7% dan 7.6%.
